```python
import pandas as pd
import matplotlib.pyplot as plt

# CSVファイル読み込み
df = pd.read_csv("emdat.csv")

# カラム名確認
print("CSVのカラム一覧:", df.columns)

# 列名を柔軟に対応
def find_column(possible_names):
    for name in df.columns:
        if name.lower() in [p.lower() for p in possible_names]:
            return name
    return None

year_col = find_column(["Year", "year", "YEAR"])
deaths_col = find_column(["Deaths", "deaths", "DEATHS"])

# 年代別死者数グラフ
if year_col and deaths_col:
    deaths_by_year = df.groupby(year_col)[deaths_col].sum()

    plt.figure(figsize=(10,6))
    deaths_by_year.plot(kind="bar", color="red")
    plt.title("年代別 死者数")
    plt.xlabel("Year")
    plt.ylabel("Number of Deaths")
    plt.tight_layout()
    plt.show()
else:
    print("⚠️ 'Year' または 'Deaths' 列が見つかりません。")
```

```python
import pandas as pd
import folium

# CSVファイル読み込み
df = pd.read_csv("emdat.csv")

# カラム名確認
print("CSVのカラム一覧:", df.columns)

# 列名を柔軟に対応
def find_column(possible_names):
    for name in df.columns:
        if name.lower() in [p.lower() for p in possible_names]:
            return name
    return None

lat_col = find_column(["Latitude", "Lat", "latitude", "lat"])
lon_col = find_column(["Longitude", "Lon", "longitude", "lon"])
year_col = find_column(["Year", "year"])
deaths_col = find_column(["Deaths", "deaths"])

# 地図の中心を平均値で設定
m = folium.Map(location=[df[lat_col].mean(), df[lon_col].mean()], zoom_start=4)

# 各災害をマッピング
for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row[lat_col], row[lon_col]],
        radius=6,
        popup=f"Year: {row.get(year_col,'N/A')}, Deaths: {row.get(deaths_col,'N/A')}",
        color="red",
        fill=True,
        fill_color="red",
        fill_opacity=0.7
    ).add_to(m)

# 地図を保存（ブラウザで開ける）
m.save("disaster_map.html")
print("✅ 災害被災マップを disaster_map.html に保存しました。ブラウザで開いてください。")
```
